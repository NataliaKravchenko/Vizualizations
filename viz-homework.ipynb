```python
df=pd.read_csv(r"C:\Users\Nat\Downloads\housing (1).data",
                      sep='\\s+')
```

```python
df.columns = ['crime_rate','<25,000 sq.ft','non-ret_buz','riv_tract','nox_conc','num_rooms','age_built<1940',
              'dist_empl','acc_highway','prop_tax','pup-teach_ratio','black','low_pop','price']
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,8))

axes[0][0].plot(df['crime_rate'], color='b')
axes[0][0].set_title('Crime rate in Boston suburbs')
axes[0][0].set_ylabel('Crime rate')
axes[0][1].plot(df['price'], color='y')
axes[0][1].set_title('House price in Boston suburbs')
axes[0][1].set_ylabel('Price')
axes[1][0].plot(df['acc_highway'], color='g')
axes[1][0].set_title('Access to highways')
axes[1][0].set_ylabel('Access')
axes[1][1].plot(df['pup-teach_ratio'], color='r')
axes[1][1].set_title('Pupils to teachers ratio')
axes[1][1].set_ylabel('Ratio')
```

```python
#a. Part 2.
fig, axes = plt.subplots(2, 2, figsize=(8,8))

axes[0][0].scatter(df['price'], df['dist_empl'], color='b')
axes[0][0].set_title('Price vs dist_emp')
axes[0][0].set_xlabel('Price')
axes[0][0].set_ylabel('dist_empl')

axes[0][1].scatter(df['acc_highway'], df['dist_empl'], color='g')
axes[0][1].set_title('acc_highway vs dist_emp')
axes[0][1].set_xlabel('Acc_highway')
axes[0][1].set_ylabel('dist_empl')

axes[1][0].scatter(df['non-ret_buz'], df['dist_empl'], color='y')
axes[1][0].set_title('non-ret_buz vs dist_empl')
axes[1][0].set_xlabel('non-ret_buz')
axes[1][0].set_ylabel('dist_empl')

axes[1][1].scatter(df['nox_conc'], df['dist_empl'], color='g')
axes[1][1].set_title('nox_conc vs dist_empl')
axes[1][1].set_xlabel('nox_conc')
axes[1][1].set_ylabel('dist_empl')
```

```python
access = df[df['riv_tract'] == 0]
no_access = df[df['riv_tract'] == 1]
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1, projection='3d')
line1 = axes.scatter(access['dist_empl'], access['price'], access['nox_conc'])
line2 = axes.scatter(no_access['dist_empl'], no_access['price'], no_access['nox_conc'])
axes.legend((line1, line2), ('access', 'no_access'))
axes.set_xlabel('dist_empl')
axes.set_ylabel('price')
axes.set_zlabel('nox_conc')
```
